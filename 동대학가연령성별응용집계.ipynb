```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path='E:\\'
os.listdir(path)
```

```python
UPJONG_Df=pd.read_csv(os.path.join(path, '블록_성별연령대별_201901.txt'),encoding='utf-8', sep='`')
UPJONG_Df
```

```python
UPJONG_Df=UPJONG_Df.drop(UPJONG_Df.columns[[0,2,4,6,8,10,12,14]],axis=1)
UPJONG_Df
```

```python
UPJONG_Df.rename(columns={'서울시민업종코드':'UPJONG_CD'},inplace=True)
UPJONG_Df
```

```python
path2='E:\\063. 서울시민의 업종별 카드소비 패턴 데이터\\2. 파일데이터'
os.listdir(path)
```

```python
blockDf=pd.read_excel(os.path.join(path2, '블록코드정리표.xlsx'))
blockDf
```

```python
blockDf2=pd.DataFrame({
    'BLCK_CD':blockDf['BLCK_CD,C,6'],
    'BLCK':blockDf['BLCK'],
    'UNVTW_AT':blockDf['UNVTW_AT,C,1'],
    'RASTSP_AT':blockDf['RASTSP_AT,C,1']
})
blockDf2
```

```python
sectorCodeDf=pd.read_csv(os.path.join(path, '서울시민 카드소비 업종코드.csv'), encoding='utf-8')
sectorCodeDf['UPJONG_CD'] = sectorCodeDf['UPJONG_CD'].str.upper()
sectorCodeDf.head()
```

```python
UPJONG_Df=pd.merge(UPJONG_Df, sectorCodeDf, how='left', on='UPJONG_CD')
UPJONG_Df.info()
```

```python
UPJONG_Df
```

```python
UPJONG_Df.rename(columns={'고객주소블록코드':'BLCK_CD'},inplace=True)
UPJONG_Df
```

```python
dataForRank=pd.merge(UPJONG_Df, blockDf2, how='left', on='BLCK_CD')
dataForRank
```

```python
sac_df=dataForRank.pivot_table(index=['BLCK','UNVTW_AT','RASTSP_AT','성별','연령대별','CLASS1'],values=['카드이용금액계','카드이용건수계'],aggfunc='sum')
sac_df
```

```python
dataForRank2=sac_df.reset_index()
dataForRank2
```

```python
dataForRank2['카드이용건수계순위']=dataForRank2.groupby(['BLCK','UNVTW_AT','연령대별','성별'])['카드이용건수계'].rank(ascending=False)
dataForRank2
```

```python
dataForRank2['카드이용금액계순위']=dataForRank2.groupby(['BLCK','UNVTW_AT','연령대별','성별'])['카드이용금액계'].rank(ascending=False)
dataForRank2
```

```python
dataForRank3=dataForRank2.drop(['카드이용건수계','카드이용금액계'],axis=1)
dataForRank3
```

```python
dataForRank3.to_csv('동대학가연령성별응용집계.csv')
```
